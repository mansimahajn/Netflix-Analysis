```python
# Prepare the data
df1 = data[['type', 'release_year']]
df1 = df1.rename(columns={"release_year": "Release Year", "type": "Type"})
df2 = df1.groupby(['Release Year', 'Type']).size().reset_index(name='Total Count')

# Filter the data to show only from the year 2000 onwards
df2 = df2[df2['Release Year'] >= 1995]

# Pivot the data to have separate columns for Movies and Shows
df_pivot = df2.pivot(index='Release Year', columns='Type', values='Total Count').fillna(0)

# Create stack plot
fig = plt.figure(figsize=(9, 4))
plt.stackplot(df_pivot.index, df_pivot['Movie'], df_pivot['TV Show'], labels=['Movies', 'TV Shows'], colors=["#b20710", "#221f1f"])

# Add title and labels
plt.title('Number of Movies and TV Shows Released Over 25 Years ', fontsize=15, fontfamily='serif')

plt.ylabel('Total Count', fontsize=12, fontfamily='serif')

# Add labels for Movies and TV Shows
fig.text(0.15, 0.75, "Movie", fontweight="bold", fontfamily='serif', fontsize=15, color='#b20710')
fig.text(0.23, 0.75, "|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.25, 0.75, "TV Show", fontweight="bold", fontfamily='serif', fontsize=15, color='#221f1f')


# Show plot
plt.show()
```

<h3>Sentiment Analysis of Netflix Content<h3>

```python
# Predefined lists of positive and negative words
positive_words = ['great', 'fantastic', 'wonderful', 'excellent', 'exciting', 'thrilling', 'masterpiece', 'outstanding', 'superb', 'captivating']
negative_words = ['average', 'terrible', 'bad', 'boring', 'dull', 'poor', 'weak', 'mediocre', 'forgettable', 'awful', 'uninteresting']

# Function to determine sentiment
def determine_sentiment(description):
    description_words = description.lower().split()
    positive_count = sum(1 for word in description_words if word in positive_words)
    negative_count = sum(1 for word in description_words if word in negative_words)
    if positive_count > negative_count:
        return 'Positive'
    elif negative_count > positive_count:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment function
df3 = data[['release_year', 'description']]
df3 = df3.rename(columns={'release_year': 'Release Year', 'description': 'Description'})
df3['Sentiment'] = df3['Description'].apply(determine_sentiment)

# Group by 'Release Year' and 'Sentiment' and count occurrences
df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Count')
df3 = df3[df3['Release Year'] > 2005]

# Plotting using matplotlib
fig, ax = plt.subplots()

sentiments = df3['Sentiment'].unique()
colors = {'Positive': '#E50914', 'Neutral': '#221F1F', 'Negative': '#B81D24'}  # Netflix color palette

for sentiment in sentiments:
    subset = df3[df3['Sentiment'] == sentiment]
    ax.bar(subset['Release Year'], subset['Total Count'], label=sentiment, color=colors[sentiment])

ax.set_title("Sentiment Analysis of Content on Netflix")
ax.set_xlabel("Release Year")
ax.set_ylabel("Total Count")
ax.legend(title='Sentiment')

plt.show()
```
